```python
import pandas as pd

with open('HWtrain.data.txt', 'r') as f:
    data_train = f.readlines()
    
data_train[0]
```

```python
#wine quality가 1,2,3 으로 나뉘어져 있으므로 이를 하나의 label이라고 생각하고 풀었습니다.
column_name = ['label']
for i in range(1,14):
    col_name = 'col_'+str(i)
    column_name.append(col_name)
print(column_name)
```

```python
#strip과 split을 사용해 data_train을 알맞은 형식으로 변경
df = pd.DataFrame([x.strip().split(',') for x in data_train], columns = column_name)

df
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

x = df.values[:,1:14]

x
```

```python
#정규화 진행하고 확인하기
stdscalar = StandardScaler()
stdscalar.fit(x)
x = stdscalar.transform(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_x = pd.DataFrame(x,columns=feat_cols)

normalised_x
```

```python
from sklearn.decomposition import PCA
pca_wine = PCA(n_components=2)

print(pca_wine.fit(x))

principalComponents_Wine = pca_wine.transform(x)

principal_wine_Df = pd.DataFrame(data = principalComponents_Wine
             , columns = ['principal component 1', 'principal component 2'])

principal_wine_Df
```

### (a)	When we use the first two principle components, how much information of the total do we use?

```python
variance_ratio = pca_wine.explained_variance_ratio_
total_variance_ratio = np.sum(variance_ratio)

total_variance_ratio
```

PC 1과 2의 분산의 비율을 합쳐서 그 합이 0.54임을 확인할 수 있었다. 따라서 전체 데이터의 55프로 정도를 사용하고 있다고 해석할 수 있다.

### (b)	Find the top-three original variables among the 13 that contributed to the construction of the first principle component.

```python
pca_wine.components_[0]
```

```python
top_three_indices = pca_wine.components_[0].argsort()[::-1][:3]

top_three_indices+1
```

PC1의 weight vector를 살펴보면 어떤 variable이 영향을 크게 미쳤는지 알 수 있다. 1번째, 11번째, 6번째 variable이 가장 큰 영향을 미쳤음을 알 수 있다.

우선 scoring vector 들을 시각화했다.

```python
principal_wine_Df
```

```python
label = df['label']
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Wine Dataset",fontsize=20)
targets = list(np.unique (label))
print(targets)

colors = ['r', 'g', 'b']
for target, color in zip(targets, colors):
    indicesToKeep = (label == target)
    plt.scatter(principal_wine_Df.loc[indicesToKeep, 'principal component 1']
               , principal_wine_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

Test data를 호출하여 데이터프레임으로 바꿔준다.

```python
with open('HWtest.data.txt', 'r') as f:
    data_train = f.readlines()
    
data_train[0]
```

```python
column_name = ['label']
for i in range(1,14):
    col_name = 'col_'+str(i)
    column_name.append(col_name)
print(column_name)
```

```python
#strip과 split을 사용해 data_train을 알맞은 형식으로 변경
df_test = pd.DataFrame([x.strip().split(',') for x in data_train], columns = column_name)

df_test
```

```python
#label을 제외하고 feature만 꺼내오기
WineTestInput = df_test.iloc[:,1:14]

WineTestInput
```

정규화 진행하고 train set에서 구한 weight vector를 사용해 test set의 scoring vector 구하기

```python
newPCAscores = pca_wine.transform( stdscalar.transform(WineTestInput) )
print(newPCAscores )
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Wine Dataset",fontsize=20)
targets = list(np.unique (label))

colors = ['r', 'g', 'b']
for target, color in zip(targets, colors):
    indicesToKeep = (label == target)
    plt.scatter(principal_wine_Df.loc[indicesToKeep, 'principal component 1']
               , principal_wine_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
colors = ['b','g','r','k','m']
# b : blue, g : green, r : red, k : black, m : magneta
for i,color in zip(range(5),colors):
    plt.scatter(newPCAscores[i,0], newPCAscores[i,1], marker='>', c=color, s=300 )
```
