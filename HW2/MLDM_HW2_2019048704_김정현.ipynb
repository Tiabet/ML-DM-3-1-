```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

with open('data_1.txt', 'r') as f:
    data = f.readlines()

column_name = data[0].strip().split()

print(column_name)

data_split = [x.strip().split() for x in data[1:]]

print(data_split[0])
```

```python
df = pd.DataFrame(data_split, columns = column_name)

df
```

```python
X = df[['KM', 'CC', 'HP']]
y = df['Price']

model = LinearRegression()

bootstraps = 2000

preds = np.zeros(bootstraps)
means = np.zeros(bootstraps)
stds = np.zeros(bootstraps)

for i in range(bootstraps):
    sample_indices = np.random.choice(X.shape[0], size=X.shape[0], replace=True)
    X_sample = X.iloc[sample_indices]
    y_sample = y.iloc[sample_indices]
    
    model.fit(X_sample, y_sample)
    preds[i] = model.predict(X.iloc[0:1])[0]
    
    means[i] = preds[0:i+1].mean()
    stds[i] = preds[0:i+1].std()

exp_pred = means.mean()
std_pred = stds.mean()

# Print the results
print('Expectation of the predicted value:', exp_pred)
print('Standard deviation of the predicted value:', std_pred)
```
