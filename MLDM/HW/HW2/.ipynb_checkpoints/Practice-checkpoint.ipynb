```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

with open('data_1.txt', 'r') as f:
    data = f.readlines()

column_name = data[0].strip().split()

print(column_name)

data_split = [x.strip().split() for x in data[1:]]

print(data_split[0])
```

```python
df = pd.DataFrame(data_split, columns = column_name)

df = df.iloc[0:5,]
df
```

```python
X = df[['KM', 'CC', 'HP']]
y = df['Price']

model = LinearRegression()

bootstraps = 100

preds = np.zeros(bootstraps)
means = np.zeros(bootstraps)
stds = np.zeros(bootstraps)

for i in range(bootstraps):
    sample_indices = np.random.choice(X.shape[0], size=X.shape[0], replace=False)
    
    X_sample = X.iloc[sample_indices]
    y_sample = y.iloc[sample_indices]
    model.fit(X_sample, y_sample)
    print(model.predict(X.iloc[0:4]))
```
